```python
# viewing our dataframe the first ten rows
# getting familiar with the data
import pandas as pd
df_salaries = pd.read_csv('salaries_by_college_major.csv')
df_salaries.head(10)
```

```python
# checking if there is NAN, then we see that at the last row, there is nan
df_salaries.isna().values.any()
```

```python
df_salaries.dropna(inplace=True)
```

```python
top5_Major = df_salaries.head(5)
```

```python
# query the highest starting salary paid for major, so the highest salary is $74,300
df_salaries['Starting Median Salary'].max()
```

```python
# query the lowest starting salary paid for major, so the lowest salary is $34,000
df_salaries['Starting Median Salary'].min()
```

```python
# the average salary paid under starting salary
Average_Salary = df_salaries['Starting Median Salary'].sum()/len(df_salaries['Starting Median Salary'])
Average_Salary
```

```python
# The major that receive the highest salary under Starting Median is Physician Assistant with the salary amount of $74,000
#  based on the PayScale Inc. Survey
df_salaries['Starting Median Salary'].idxmax()
df_salaries.loc[43]
```

```python
# The major that receive the lowest salary under Starting Median is Spanish with the salary amount of $34,000
df_salaries['Starting Median Salary'].idxmin()
df_salaries['Undergraduate Major'].loc[49]
df_salaries.loc[49]
```

```python
df_salaries['Mid-Career Median Salary'].idxmax()
```

```python
# The major that receive the highest salary under mid-career is Chemical Engineering with the salary amount of $107,000
df_salaries['Undergraduate Major'].loc[8]
df_salaries.loc[8]
```

```python
# The major that receive the lowest salary under mid-career is sadly Education with the salary amount of $52,000
df_salaries['Mid-Career Median Salary'].idxmin()
df_salaries['Undergraduate Major'].loc[18]
df_salaries.loc[18]
```

```python
# getting the some statistical results about the data
df_salaries.describe().round(2)
```

```python
#adding up the salary to get total salary
all_salary = df_salaries['Starting Median Salary']+ df_salaries['Mid-Career Median Salary']+ df_salaries['Mid-Career 10th Percentile Salary']+ df_salaries['Mid-Career 90th Percentile Salary']
```

```python

#adding the new column - total salary into the dataframe
Average = all_salary/4
```

```python
df_salaries.insert(6, 'Average_salary', Average)
```

```python
#showing the dataframe after adding column average salary
df_salaries
```

```python
#sorting data using total and average salary in descending order
sorted_df = df_salaries.sort_values(['Total_salary', 'Average_salary'], ascending=False)
```

```python
# showing the sorted data
sorted_df
```

```python
top5_Major_by_avgtotal_sal = sorted_df.head(5)
```

```python
top5_Major_by_avgtotal_sal
```

```python
bottom5_Major_by_avgtotal_sal = sorted_df.tail(5)
```

```python
bottom5_Major_by_avgtotal_sal
```

```python
# getting the sum of group per salary
df_salaries.groupby('Group').sum()
```

```python
# getting the average value of group per salary and make it two decimal
round(df_salaries.groupby('Group').mean(), 2)
```

```python
df_salaries.pivot_table(index='Undergraduate Major', aggfunc="mean")
```

```python
starting_sal = df_salaries.pivot_table(index='Undergraduate Major', values="Starting Median Salary", aggfunc="sum")
```

```python
starting_sal
```

```python
import matplotlib.pyplot as plt
```

```python
starting_sal.plot(kind='bar', xlabel='Undergraduate Major', ylabel='Salary',
                  title='Starting Salary per Undergraduate Major Report', figsize=(12,6))
plt.savefig('my_survey.png')
plt.show()
```

```python
starting_sal.to_excel('pivot_table.xlsx')
```

```python
sample_sal = df_salaries.pivot_table(index='Undergraduate Major', values=["Starting Median Salary", "Mid-Career 10th Percentile Salary"], aggfunc="sum")
```

```python
sample_sal.plot(kind='bar', xlabel='Undergraduate Major', ylabel='Salary',
                  title='Starting and 10th Percentile Salary per Undergraduate Major Report', figsize=(12,6))
plt.savefig('my_survey_2.png')
plt.show()
```

```python
starting_salary_top5 = top5_Major_by_avgtotal_sal.pivot_table(index='Undergraduate Major', values="Starting Median Salary", aggfunc="sum")
```

```python
starting_salary_top5
```

```python
# showing top 5 on pie chart
starting_salary_top5.plot(kind='pie', y='Starting Median Salary', title='Top 5 Undergraduate Major Under Starting Salary')
plt.savefig('my_survey_2.png')
plt.show()
```
